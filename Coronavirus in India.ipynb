# Linear Regression

```python
X=df[['Total_confirmed', 'Latitude', 'Longitude']] #Taking independent variable seperately
y= df['Cured/Discharged/Migrated']                  ##Taking dependent variable seperately
```

```python
from sklearn.model_selection import train_test_split #importing train-test from sklearn library for training and testing our model
```

```python
X_train, X_test , y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 100) #splitting data in 30-70 ratio
```

```python
from sklearn.linear_model import LinearRegression #importing linear regresssion from sklearn library
```

```python
lm= LinearRegression() # taking linear regression to lm
```

```python
lm.fit(X_train,y_train) #fitting of data into model
```

```python
pred=lm.predict(X_test)
```

```python
plt.scatter(y_test,pred)
```

```python
from sklearn import metrics
print('MAE',metrics.mean_absolute_error(y_test,pred))
print('MSE',metrics.mean_squared_error(y_test,pred))
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test,pred)))
```

**Inference**- Since our Mean squared error and Root Mean Square error is very less we can say that our model is ready for right prediction
